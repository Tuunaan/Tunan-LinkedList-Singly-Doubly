```python
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n


class LinkedList:
  def __init__(self, a):
    self.a=a
    self.head=Node(a[0], None)
    tail=self.head
    for i in range(1, len(a)):
      n= Node(a[i], None)
      tail.next=n
      tail=tail.next

  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s

  def countNode(self):
    count=0
    temp=self.head
    while temp!=None:
      count+=1
      temp=temp.next
    return count

  def nodeAt(self, idx):
    lenn=self.countNode()
    if idx<0 or idx>=lenn:
      return None
    else:
      temp=self.head
      x=0
      while temp is not None:
        if x==idx:
          return temp
        x=x+1
        temp=temp.next

  def get(self, idx):
    temp=self.head
    x=0
    while temp is not None:
      if x==idx:
        return temp.element
      x=x+1
      temp=temp.next

  def set(self, idx, elem):
    lenn=self.countNode()
    temp=self.head
    x=0
    while temp is not None:
      if x==idx:
        value= temp.element
        temp.element=elem
        return value
      elif idx<0 or idx>=lenn:
        return None
      else:
        x=x+1
        temp=temp.next

  def indexOf(self, elem):
    temp=self.head
    idx=0
    while temp is not None:
        if temp.element==elem:
          return idx
        idx=idx+1
        temp=temp.next
    return ("Invalid Element")

  def contains(self, elem):
    temp=self.head
    idx=0
    while temp is not None:
        if temp.element==elem:
          return True
        idx=idx+1
        temp=temp.next
    return False

  def copyList(self):
    copyH=None
    copyT=None
    temp=self.head
    x=0
    while temp is not None:
      new_node=Node(temp.element,None)
      if copyH==None:
        copyH=new_node
        copyT=new_node
      else:
        copyT.next=new_node
        copyT=new_node
      temp=temp.next
    return copyH

  def reverseList(self):#out_place
    revH=None
    temp=self.head
    while temp is not None:
      new_node=Node(temp.element,None)
      if(revH == None):
          revH = new_node
      else:
          new_node.next = revH
          revH = new_node
      temp = temp.next
    return revH



  def reverseList1(self):#in_place
    newHead = None
    temp= self.head
    while(temp is not None):
        nextNode = temp.next
        temp.next = newHead
        newHead = temp
        temp= nextNode
    return newHead

  def createList(self,array):
      head = None
      tail = None
      for i in range(0, len(array)):
          newNode = Node(array[i], None)
          if(head == None):
              head = newNode
              tail = newNode
          else:
              tail.next = newNode
              tail = newNode
      return head

  def insert(self, elem, idx):
    lenn=self.countNode()
    new_node=Node(elem,None)
    if idx<0 or idx>=lenn+1:
      print("Invalid Index")
    elif idx==0:
      new_node.next=self.head
      self.head=new_node
    else:
      previous_node=self.nodeAt(idx-1)
      new_node.next=previous_node.next
      previous_node.next=new_node

  def remove(self, idx):
    lenn=self.countNode()
    val=self.nodeAt(idx).element
    if idx<0 or idx>=lenn:
      return -1
    elif idx==0:
      removed_head=self.head
      self.head=self.head.next
      removed_head.next=None
    else:
      previous_node=self.nodeAt(idx-1)
      removed_node=previous_node.next
      previous_node.next=removed_node.next
      removed_node.next=None
    return val

  def rotateLeft(self):
    old=self.head
    self.head=self.head.next
    tail=self.head
    while tail.next is not None:
      tail=tail.next
    tail.next=old
    old.next=None
    return self.head

  def rotateRight(self):
    temp=self.head
    removed=None
    while temp.next.next!=None:
        temp=temp.next
        removed=temp.next
    temp.next=None
    node=Node(removed.element,self.head)
    self.head=node

```
